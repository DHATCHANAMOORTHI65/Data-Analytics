**Data Preprocessing**

```python
ele=pd.DataFrame(data)
ele
```

**Feature Enginnering**

```python
ele['Electoral_Votes']=ele['Electoral_Votes'].fillna(0)
ele['Harris_Poll (%)']=pd.to_numeric(ele['Harris_Poll (%)'])
ele['Trump_Poll (%)']=pd.to_numeric(ele['Trump_Poll (%)'])
ele['Margin']=pd.to_numeric(ele['Margin'])
ele['Electoral_Votes']=pd.to_numeric(ele['Electoral_Votes'])
ele
```

```python
def calculate_swing(row):
  if "D" in row["2020_Result"]:
    return row["Harris_Poll (%)"] -  50
  elif "R" in row["2020_Result"]:
    return row["Harris_Poll (%)"] - 50
  else:
    return row["Harris_Poll (%)"] - 50

ele['Historical_Swing'] = ele.apply(calculate_swing, axis=1)
sns.set(style="whitegrid")
```

**Visualization**

```python
plt.figure(figsize=(10,6))
sns.barplot(data=ele,x='Margin',y='State',hue='Poll_Leader',dodge=False)
plt.title('Poll Margin by State')
plt.xlabel('Margin (Harris - Trump)')
plt.ylabel('State')
plt.legend(title='Poll Leader')
plt.show()
```

```python
ev_distribution=ele.groupby("Poll_Leader")['Electoral_Votes'].sum()
plt.figure(figsize=(8,8))
plt.pie(ev_distribution,labels=ev_distribution.index, autopct='%1.1f%%',startangle=140)
plt.title('Electoral Votes distribution by Leading Candidate')
plt.show()
```

```python
plt.figure(figsize=(10,6))
ele_melted=ele.melt(id_vars=['State'], value_vars=['Harris_Poll (%)','Trump_Poll (%)'],var_name='Candidate',value_name='Poll (%)')
sns.barplot(data=ele_melted,x='Poll (%)',y='State',hue='Candidate')
plt.title('Poll Percentages by Candidate and State')
plt.xlabel('Poll (%)')
plt.ylabel('State')
plt.legend(title='Candidate')
plt.show()
```

```python
def extract_margin(result):
  if result=="N/A":
    return 0
  margin_value=result.split()[1]
  if margin_value=="<1":
    return 0.5
  return int(margin_value)

ele['2020_Result_Margin']=ele['2020_Result'].apply(extract_margin)
plt.figure(figsize=(10,6))
sns.scatterplot(data=ele,x='Margin',y='2020_Result_Margin',hue='Poll_Leader',s=100)
plt.title('Current Poll vs 2020 Election Result')
plt.xlabel('Current Poll Margin(Harris - Trump)')
plt.ylabel('2020 Result Margin (D or R)')
plt.legend(title='Poll Leader')
plt.show()
```
